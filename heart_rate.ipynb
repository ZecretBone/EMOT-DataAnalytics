```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
self_annotation_multimodal = pd.read_excel('datasets/ECG_GSR_Emotions/Self-Annotation Labels/Self-annotation Multimodal_Use.xlsx')
self_annotation_multimodal['annotation'] = 'M'
self_annotation_multimodal.rename(columns = {'V_Label':'Valence', 'A_Label':'Arousal', 'Four_Labels':'Four_Label'}, inplace = True)
self_annotation_multimodal
```

```python
dummies1 = pd.get_dummies(self_annotation_multimodal.Gender)
 
# Concatenate the dummies to original dataframe
merged1 = pd.concat([self_annotation_multimodal.iloc[:, :5], dummies1, self_annotation_multimodal.iloc[:, 6:]], axis='columns')
merged1
```

```python
self_annotation_singlemodal = pd.read_excel('datasets/ECG_GSR_Emotions/Self-Annotation Labels/Self-annotation Single Modal_Use.xlsx')
self_annotation_multimodal['annotation'] = 'S'
self_annotation_singlemodal
```

```python
dummies2 = pd.get_dummies(self_annotation_singlemodal.Male)
 
# Concatenate the dummies to original dataframe
merged2 = pd.concat([self_annotation_singlemodal.iloc[:, :5], dummies2, self_annotation_singlemodal.iloc[:, 6:]], axis='columns')
merged2
```

```python
self_annotation_frames = [merged1, merged2]
merged_dataframe = pd.concat(self_annotation_frames)
merged_dataframe
```

```python
train_d_frame = merged_dataframe.copy().drop(['Video ID', 'Happy', 'Sad', 'Fear',
                                        'Anger', 'Neutral', 'Disgust', 'Surprised', 'Four_Label', 'annotation'
                                       ], axis = 1)
# train_d_frame.drop(colummns=[c for c in train_d_frame.columns if 'Unnamed' in c], )
for x in train_d_frame.columns:
    if 'Unnamed' in x:
        train_d_frame.drop(columns=x, inplace=True)

train_d_frame
```

```python
features = ['Age', 'Female', 'Male', 'Arousal level', 'Arousal', 'Valence']
cl = { 'LA':0, 'HA':1}
train_d_frame['Arousal'] = train_d_frame['Arousal'].map(cl)
X = train_d_frame[features]
y = train_d_frame.iloc[:, 10]
```

```python
#Preparing the model
from sklearn.model_selection import train_test_split, GridSearchCV, KFold
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 99)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()

cv = KFold(n_splits=10, random_state=1, shuffle=True)

grid_params_knn = { 'n_neighbors' : range(1,20)}

gs = GridSearchCV(KNeighborsClassifier(), grid_params_knn, cv=cv)
gs.fit(X_train, y_train)
gs.best_params_
gs.score(X_train, y_train)
```
